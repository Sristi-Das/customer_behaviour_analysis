```python
df.isnull().sum() #summarises the number of null values in each column
#As we can see there are 37 null values for Review Rating. We must decide how to resolve the issue of these null values. 
#We may decide to delete the rows with these null values or impute them with an appropriate value.
#We may decide to impute the null values with the mean or the median of the column.
#Mean may not be appropriate as it may be skewed by outliers.
#Median may be appropriate as it may be robust to outliers.
#The null values may be imputed with the overall median, however this may introduce bias into the data.
#The reason for this is that the reviews can be segmented by the category (such as clothing and footwear), and the ratings within each category may vary significantly.
#Therefore, if all the null values are imputed with the overall medium for the column, this may suggest that the footwear review ratings should have the same median for the null values as the clothing review rating.
#This may introduce bias into the dataset.
#Instead the null values for the Review Ratings in each category may be imputed with the the median review rating for each category.
#Therefore if a clothing item is missing a review rating, it is imputed with the median review rating for clothing specifically, instead of the overall median or the median review rating of footwear.
```

```python
df.groupby('Category') 
#df.groupby functions similary to GROUP BY in SQL. It is necessary to perform aggregate functions on certain subsets or groups of data.
#For example, in SQL, to determine the mean of review ratings alone with other columns you would GROUP BY the other columns then execute AVG(Review Ratings)
#Likewise in Python we can group data by category and perform aggregate functions on them.
```

```python
df['Review Rating'].describe() #Describes summary statistics of just Review Ratings.
```

```python
df.groupby('Category')['Review Rating'].describe() 
#Groups the entire dataset by the category column. Selects the Review Rating column from the grouped dataset. Then describes the summary statistics of Review Ratings grouped by category.
```

```python
df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
#Group the dataset by category. Select the Review Rating Column. Transform this column (temporarily named x) by imputing all null values with the median values of the column.
#The lambda and x is mostly to define a temporary short function, and may be considered syntax related for the purposes of the transformation.
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) 
#Fill all null values in the Review Rating column by the median Review Rating associated with the corresponding category for the row.
```

```python
df['age'].head(5) #There may be another column associated with age. Organising the ages into age groups such as Young Adult, Adult, Middle-Agged, and Senior
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior'] #Create a list of age group labels.
```

```python
pd.qcut(df['age'], q = 4, labels = labels) #Selects the age column in the dataframe. Sorts the ages into four quartile groups (25%, 50%, 75%, 100%). Then assigns each quartile group with the corresponding label in the labels list earlier. 
```

```python
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels) #Assigns age group labels for each age.
df[['age', 'age_group']].head(10) #Displays the upper 10 rows of age and age_group.
#Age group labels seem appropriate.
```

```python
df['frequency_of_purchases'].unique() #Displays a list of unique values in the frequency_of_purchases column in the dataframe.
```

```python
#It may be of utility to create a new column to represent the categorical values in the frequency_of_purchases column as numerical values that represent the number of days between purchases.
frequency_mapping = {'Fortnightly': 14,
                     'Weekly': 7,
                     'Annually': 365,
                     'Quarterly': 90,
                     'Bi-Weekly': 14,
                     'Monthly': 30,
                     'Every 3 Months': 90} #Creates a dictionary with numerical assignments (in days) for every unique purchase frequency type. Some frequency types are equivalent such as quarterly and Every 3 Months (equivalent to approximately 90 days).
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping) #Creates a new column called purchase_frequency_days. This column contains the numerical representation (in days) of the frequency_of_purchases column in the same row.
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10) #Displays the upper 10 rows of both 'frequency_of_purchases' and 'purchase_frequency_days' columns.
#The numerical assignments for each type of frequency_of_purchases appears appropriate.
```

```python
#This code may assist in establishing connection with PostgreSQL.
from sqlalchemy import create_engine #This function may be required to connect pandas to the PostgreSQL database table.

#Step 1: Connect to PostgreSQL
#Replace placeholders with your actual details
#To confirm actual details, you can access PGAdmin4 -> Server -> Right Click on PostgreSQL 18 -> Properties -> Connection
username ='postgres' #default user
password = 'password' #the password you set during the installation of PostgreSQL
host = 'localhost' #localhost may be utilised if operating locally
port = '5432' #default PostgreSQL port
database = 'customer_behaviour' #the database created in pgAdmin4, named customer_behaviour

engine = create_engine(f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}') 
#Creates an engine object with an f string. This f string is a URL to the database.
#postgresql+psycopg2 may suggest that the postgresql dialect may be utilised. psycopg2 may be the Python driver, which may be a library that communicates with PostgreSQL.
#engine is then the variable that may be utilised in pandas functions like df.to_sql("table_name", engine) to execute queries or supply data.

#Step 2: Load Dataframe into PostgreSQL
table_name = 'customer' #Any table name may be elected.
df.to_sql(table_name, engine, if_exists = 'replace', index = False)
#df.to_sql Sends the rows of the pandas dataframe to the database as a table.
#table_name is the name of the table in the PostgreSQL database
#The engine contains the PostgreSQL connection details and may be utilised as the connection destination.
#if_exists = 'replace' may replace any existing table_name with a recreated table_name with rows from the dataframe.
#index = False prevents pandas from generating a new column in the SQL table for the dataframe's index. This may ensure that only the original dataframe's columns may be utilised.

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
#May print: Data successfully loaded into table 'customer' in database 'customer_behaviour'.
#Verify if the code may have executed successfully in pgAdmin4. Servers -> PostgreSQL 18 -> Databases -> customer_behaviour -> Schemas -> public -> Tables -> customer.
#customer is the table that may be used to execute SQL queries.
```
